```python
import networkx as nx
import matplotlib.pyplot as plt
def visualize(visited):
    graph=[]
    for i in range(len(visited)):
        if i+1<len(visited): 
            temp=(visited[i],visited[i+1])
            graph.append(temp)
    g=nx.Graph()
    g.add_edges_from(graph)
    nx.draw_networkx(g)
    plt.show()
def recdfs(graph,source,goal):
    print("\nRDFS:")
    visited=[]
    def stacksearch(graph,v,goal):
        if v not in visited:
            print(v,end="")
            visited.append(v)
            if v!=goal: print("->",end="")
        if v==goal: 
            visualize(visited)
            return 
        for i in graph[v]:
            if i not in visited: stacksearch(graph,i,goal)
    stacksearch(graph,source,goal)
#DFS Using Recursive Function
def dfs(graph,source,goal):
    print("\nDFS:")
    visited=[]
    stack=[source]
    while stack!=[]:
        v=stack.pop()
        if v not in visited: 
            print(v,end="")
            visited.append(v)
            if v!=goal: print("->",end="")
        if v==goal: 
            visualize(visited)
            break
        l=graph[v]
        l=reversed(l)
        for i in l:
            if i not in visited: stack.append(i)
def bfs(graph,source,goal):
    print("\nBFS:")
    visited=[]
    queue=[source]
    while queue!=[]:
        v=queue.pop(0)
        if v not in visited: 
            print(v,end="")
            visited.append(v)
            if v!=goal: print("->",end="")
        if v==goal: 
            visualize(visited)
            break
        for i in graph[v]:
            if i not in visited: queue.append(i)
def dls(graph,source,goal,limit):
    print("\nDLS:")
    visited=[]
    def depthlimitedsearch(graph,v,goal,limit):
        if v not in visited and limit>=0:
            visited.append(v)
            if v==goal: 
                print(v,end="")
                return True
            else: print(v,end="->")
            if limit>0:
                for i in graph[v]:
                    if i not in visited:
                        if depthlimitedsearch(graph,i,goal,limit-1):
                            return True
        return False
    result=depthlimitedsearch(graph,source,goal,limit)
    if result==False: print("\nGoal Not found, LIMIT EXCEED!!")
    visualize(visited)

graph=dict()
g=nx.Graph()
n=int(input("Enter the number of vertex in a graph:"))
for i in range(n):
    v=input("Enter the Vertex:")
    graph[v]=list()
if n!=0:
    n=int(input("Enter the number of edges in a graph:"))
    for i in range(n):
        v=input("From:")
        u=input("To:")
        graph[v].append(u)
        graph[u].append(v)
        g.add_edge(u,v)
    nx.draw_networkx(g)
    plt.show()
    source=input("Enter the source vertex:")
    goal=input("Enter the goal vertex:")
    dfs(graph,source,goal)
    recdfs(graph,source,goal)
    bfs(graph,source,goal)
    limit=int(input("Enter the Depth Limit:"))
    dls(graph,source,goal,limit)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
def visualize(visited):
    graph=[]
    for i in range(len(visited)):
        if i+1<len(visited): 
            temp=(visited[i],visited[i+1])
            graph.append(temp)
    g=nx.Graph()
    g.add_edges_from(graph)
    nx.draw_networkx(g)
    plt.show()
def iterative_depthlimitedsearch(graph,source,goal,limit):
    print("\nIterative Deepening Search:")
    visited=[]
    def depthlimitedsearch(graph,v,goal,limit):
        if v not in visited and limit>=0:
            visited.append(v)
            if v==goal: 
                print(v,end="")
                return True
            else: print(v,end="->")
            if limit>0:
                for i in graph[v]:
                    if i not in visited:
                        if depthlimitedsearch(graph,i,goal,limit-1):
                            return True
        return False
    for dlimit in range(limit+1):
        visited.clear()
        print("\nTill depth "+str(dlimit)+":")
        if depthlimitedsearch(graph,source,goal,dlimit):
            print(" Goal state reached at Depth "+str(dlimit))
            visualize(visited)
            break
        else: 
            print(" Goal state not found at Depth "+str(dlimit))
            visualize(visited)

graph=dict()
g=nx.Graph()
n=int(input("Enter the number of vertex in a graph:"))
for i in range(n):
    v=input("Enter the Vertex:")
    graph[v]=list()
if n!=0:
    n=int(input("Enter the number of edges in a graph:"))
    for i in range(n):
        v=input("From:")
        u=input("To:")
        graph[v].append(u)
        graph[u].append(v)
        g.add_edge(u,v)
    nx.draw_networkx(g)
    plt.show()
    source=input("Enter the source vertex:")
    goal=input("Enter the goal vertex:")
    limit=int(input("Enter the Depth Limit:"))
    iterative_depthlimitedsearch(graph,source,goal,limit)
```

```python
#Corrected Uninformed Cost Search for Optimal Path
import networkx as nx
import matplotlib.pyplot as plt
def weight_graph(graph,pair,weight,cost,status,visit):
    v,p=pair
    if status==True:
        for v1,w in graph[v]:
            if p==v1:
                weight+=w
                if weight<=cost: 
                    visit.append((v,p,w))
                    return [True,weight,visit]
                else:
                    weight-=w
                    return [None,weight,visit]
    else:
        for v1,w in graph[v]:
            if p==v1:
                weight-=w
                for i in range(len(visit)):
                    if v==visit[i][0] and p==visit[i][1] and i!=0: visit.pop(i)
                return [True,weight,visit]
def visualize(graph,visited,cost):
    g=nx.Graph()
    count=0
    weight=0
    visited.pop(0)
    visit=[]
    status=True 
    for pair in visited:
        list=weight_graph(graph,pair,weight,cost,status,visit)
        if list[0]==True: 
            weight=list[1]
            visit=list[2]
            count+=1
            status=True
        else:
            weight=list[1]
            visit=list[2]
            status=False
            list=weight_graph(graph,visited[count-1],weight,cost,status,visit)
            if list[0]==True: 
                weight=list[1]
                visit=list[2]
                status=True
    for v,u,w in list[2]:
        g.add_edge(v,u,weight=w)
    pos=nx.spring_layout(g)
    weight_label=nx.get_edge_attributes(g, "weight")
    nx.draw(g,pos,with_labels=True)
    nx.draw_networkx_edge_labels(g,pos,weight_label)
def uniformed_cost_search(graph,source,goal):
    print("Uniformed Cost Search:")
    visited=[]
    visit=[]
    queue=[(0,source,None)]
    while queue!=[]:
        queue.sort()
        weight,vertex,parent=queue.pop(0)
        if vertex not in visit:
            visit.append(vertex)
            visited.append((vertex,parent))
            if vertex==goal: 
                print(vertex,end="")
                visualize(graph,visited,weight)
                break
            print(vertex,end="->")
        for adj,w in graph[vertex]:
            if adj not in visit: queue.append((weight+w,adj,vertex))
graph=dict()
n=int(input("Number of Vertex in a graph:"))
for i in range(n):
    v=input("Vertex:")
    graph[v]=[]
if n!=0:
    G=nx.Graph()
    n=int(input("Number of Edges in a graph:"))
    for i in range(n):
        v=input("From:")
        u=input("To:")
        w=int(input("Weight:"))
        graph[v].append((u,w))
        graph[u].append((v,w)) #Bidirectional graph
        G.add_edge(v,u,weight=w)
    pos=nx.spring_layout(G)
    weight_label=nx.get_edge_attributes(G, "weight")
    nx.draw(G,pos,with_labels=True)
    nx.draw_networkx_edge_labels(G,pos,weight_label)
    plt.show()
    source=input("Source:")
    goal=input("Goal:")
    uniformed_cost_search(graph,source,goal)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
def visualize(visited):
    graph=[]
    temp=[]
    for i in visited.keys():
            if visited[i]==True: temp.append(i)
    for i in range(len(temp)):
        if i+1<len(temp):
            t=(temp[i],temp[i+1])
            graph.append(t)
    g=nx.Graph()
    g.add_edges_from(graph)
    nx.draw_networkx(g)
    plt.show()
def bidirectional_search(graph,source,goal):
    print("Bidirectional Search")
    visited_s={}
    visited_g={}
    source_queue=[source]
    goal_queue=[goal]
    found=False
    while source_queue!=[] and goal_queue!=[]:
        v=source_queue.pop(0)
        u=goal_queue.pop(0)
        visited_s[v]=False
        visited_g[u]=False
        for status,vertex in graph[v]: 
            if vertex not in visited_s.keys() and status==1:
                visited_s[v]=True
                source_queue.append(vertex)
        for status,vertex in graph[u]: 
            if vertex not in visited_g.keys() and status==0:
                visited_g[u]=True
                goal_queue.append(vertex)
        for source in source_queue:
            if source in goal_queue:            
                visited_s[source]=True
                for i in reversed(visited_g.keys()):
                    visited_s[i]=visited_g[i]
                visualize(visited_s)
                found=True
                break
        if found: break
graph=dict()
g=nx.Graph()
n=int(input("Enter the number of vertex in a graph:"))
for i in range(n):
    v=input("Enter the Vertex:")
    graph[v]=list()
if n!=0:
    n=int(input("Enter the number of edges in a graph:"))
    for i in range(n):
        v=input("From:")
        u=input("To:")
        graph[v].append((1,u))
        graph[u].append((0,v))
        g.add_edge(u,v)
    nx.draw_networkx(g)
    plt.show()
    source=input("Enter the source vertex:")
    goal=input("Enter the goal vertex:")
    bidirectional_search(graph,source,goal)
```
